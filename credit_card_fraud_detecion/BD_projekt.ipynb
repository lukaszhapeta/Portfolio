Confusion matrix (kolejność ze względu na bibliotekę sklearn):

$$
\begin{bmatrix}
    \text{TN} & \text{FP} \\
    \text{FN} & \text{TP}
\end{bmatrix}
$$

gdzie:
- TN - true negative, czyli klasa 0, której przypisaliśmy klasę 0
- FP - false positive, czyli klasa 0, której przypisaliśmy klasę 1
- FN - false negative, czyli klasa 1, której przypisaliśmy klasę 0
- TP - true positive, czyli klasa 1, której przypisaliśmy klasę 1

(Tutaj zaznaczę, klasa 1 - oszustwo, klasa 0 - poprawna transakcja)

Zależy nam na zminimalizowaniu pomyłek.

Załadowanie funkcji

```python
# Wykres krzywej ROC

def plot_roc(y_test, predictions, optimal = False):
    cmap_custom = ListedColormap(['#101010', '#FFFD47'])
    fig, ax_2 = plt.subplots()
    fpr, tpr, thresholds = metrics.roc_curve(y_test, predictions)
    roc_auc = metrics.auc(fpr, tpr)
    ax_2.plot(fpr, tpr, color='#FFFD47', label='Krzywa ROC (AUC = {:.2f})'.format(roc_auc))
    ax_2.set_facecolor('#101010')
    fig.set_facecolor('#101010')
    
    ax_2.spines['bottom'].set_color('white')
    ax_2.spines['top'].set_color('white')
    ax_2.spines['right'].set_color('white')
    ax_2.spines['left'].set_color('white')
    ax_2.tick_params(axis='x', colors='white')
    ax_2.tick_params(axis='y', colors='white')
    ax_2.set_xlabel('Swoistość', color = "white")
    ax_2.set_ylabel('Czułość', color = "white")
    ax_2.plot([0, 1], [0, 1], linestyle='--', color='white', label='Klasyfikator losowy', linewidth = 1)
    
    # Dorysowujemy linię
    if optimal:
        ax_2.plot([0, 0.016], [1, 0.78], linestyle='--', color='skyblue', label='Najkrótszy dystans do punktu (0,1)', linewidth = 1)
        ax_2.text(0.05, 0.75, 'Próg klasyfikacji = 0.026', color='skyblue', fontsize='small', ha='left', va='bottom')

        ax_2.legend(loc='lower right', fontsize='small', frameon=True, facecolor='white', edgecolor='white')
    

    plt.show()
    return fpr, tpr, thresholds

# Wykres confusion matrix

def plot_matrix(y_test, predictions):
    cmap_custom = ListedColormap(['#101010', '#FFFD47'])
    cm = metrics.confusion_matrix(y_test, predictions)
    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=None)
    conf_matrix = cm
    total = sum(conf_matrix.ravel())
    tn, fp, fn, tp = conf_matrix.ravel()
    accuracy = (tn + tp)/total
    precision = tp/(tp+fp)
    recall = tp / (tp + fn)
    specificity = tn / (tn + fp)
    f1_score = 2*(precision*recall)/(precision + recall)
    print("Suma pomyłek (FP + FN):", fp + fn)
    print("Dokładność:", round(accuracy, 3))
    print("Precyzja:", round(precision, 3))
    print("Czułość:", round(recall, 3))
    print("Swoistość:", round(specificity, 3))
    print("Wyznacznik F:", round(f1_score, 3))
    fig, ax_1 = plt.subplots()
    cm_display.plot(colorbar = False, cmap=cmap_custom, ax=ax_1)
    cm_display.figure_ = fig
    for i in range(len(cm)):
        for j in range(len(cm[0])):
            if ((i == 0) & (j == 0)):
                ax_1.text(j, i, str(cm[i, j]), ha='center', va='center', color='black')
            else:
                ax_1.text(j, i, str(cm[i, j]), ha='center', va='center', color='white')
    fig.set_facecolor('#101010')
    ax_1.set_xlabel('')
    ax_1.set_ylabel('')
    ax_1.spines['bottom'].set_color('white')
    ax_1.spines['top'].set_color('white')
    ax_1.spines['right'].set_color('white')
    ax_1.spines['left'].set_color('white')
    ax_1.tick_params(axis='x', colors='white')
    ax_1.tick_params(axis='y', colors='white')

    plt.show()
    
    
# Funkcja znajdująca próg dla najmniejszej sumy pomyłek

def ustal_prog(potencjalne_progi, model):
    najlepszy_prog = min(potencjalne_progi)
    przypisane_klasy = [0 if element < najlepszy_prog else 1 for element in model.predict(x_test).to_numpy()]
    tn, fp, fn, tp = metrics.confusion_matrix(y_test, przypisane_klasy).ravel()
    najmniejsza_suma_pomylek = fp + fn
    for prog in potencjalne_progi:
        przypisane_klasy = [0 if element < prog else 1 for element in model.predict(x_test).to_numpy()]
        tn, fp, fn, tp = metrics.confusion_matrix(y_test, przypisane_klasy).ravel()
        aktualna_suma_pomylek = fp+fn
        if aktualna_suma_pomylek < najmniejsza_suma_pomylek:
            najmniejsza_suma_pomylek = aktualna_suma_pomylek
            najlepszy_prog = prog

    print("Najlepszy próg dla najmniejszej sumy pomyłek:", round(najlepszy_prog, 3))
```
